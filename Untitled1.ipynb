```python
import websocket, json
```

```python
symbol = "btcusdt"
interval = "1m"
socket = f'wss://stream.binancefuture.com/ws/{symbol}@kline_{interval}'
obsocket = f'wss://stream.binancefuture.com/ws/{symbol}@depth'
```

```python
def on_message(ws,message):
    msg = json.loads(message)
    if (msg['k']['x']):
        print('----')
        print(msg['k']['c'])
def on_close(ws):
    print("Cerrado")
ws = websocket.WebSocketApp(socket, on_message= on_message, on_close = on_close)
```

```python
ws.run_forever()
```

```python
def on_message(ws,message):
    msg = json.loads(message)
    print(msg['a'])
def on_close(ws):
    print("Cerrado")
    
ob_socket = websocket.WebSocketApp(obsocket, on_message= on_message, on_close = on_close)
```
